# Κώδικας visualization


````
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[~grid.grid] = [0, 0, 0]
        image[grid.grid] = [255, 255, 255]
        # Use this to treat 1/True as obstacles
        # image[grid.grid] = [0, 0, 0]
        # image[~grid.grid] = [255, 255, 255]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)
    
    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

````
%matplotlib inline
import numpy as np
from queue import LifoQueue
import  random 
import matplotlib.pyplot as plt


class Maze:
    def __init__(self, N, S, F):

      """
      N: integer that indicates the size of the NxN grid of the maze
      S: pair of integers that indicates the coordinates of the starting point (S)
      F: pair of integers that indicates the coordinates of the finish point (F)
      You can add any other parameters you want to customize maze creation (e.g. variables that
      control the creation of additional paths)
      """

      assert N > 2

      ## Make sure start and end are within the grid

      assert S[0] < N-1
      assert S[1] < N-1
      assert F[0] < N-1
      assert F[1] < N-1

      assert S[0] > 0
      assert S[1] > 0
      assert F[0] > 0
      assert F[1] > 0

      # Add here any additional constraints your implementation may have

      self.N = N
      self.S = S
      self.F = F

      # Grid initialized with obstacles (array of 0/False)
      # 1/True indicates available cells
      self.grid = np.zeros((N, N), dtype=bool)
      
      ## YOUR CODE HERE      
      self.grid[1,1]=True #κάνουμε το κελί (1,1) ελεύθερο
      
      def GetNearCells(maze, N, x, y, value): #βρίσκει τα γειτονικά κελία που είναι είτε εμπόδια (Value=False) είτε ελεύθερα (Value=True) δωσμένου ενός κελιού
        cells = [(x+2,y),(x-2,y),(x,y+2),(x,y-2)] #όλα τα πιθανά κελία
        new = []
        for ix,iy in cells:
          if ix > 0 and ix < N-1 and iy > 0 and iy < N-1 and maze[ix,iy] == value: #έλεγχος αν είναι εντός ορίων και ότι έχει την κατάλληλη ιδιότητα
            new.append((ix,iy)) #προσθήκη στην λίστα που θα επιστραφεί
        return new 

      def GetFrontier(maze, N, x, y): #βρίσκει τα γειτονικά κελιά που είναι εμπόδια
        return GetNearCells(maze, N, x, y, False)

      def GetNeighbours(maze, N, x, y): #βρίσκει τα γειτονικά κελιά που είναι ελεύθερα κελιά
        return GetNearCells(maze, N, x, y, True) 
      
      def mergefrontiers(maze, frontier, newFrontier): #ανενώνει το frontier
        new = []
        for ix,iy in frontier: #ελέγχει τα κελιά που είναι στο παλιό frontier
          if maze[ix,iy] == 0:  #αν είναι εμπόδιο τότε μπαίνει στο νέο frontier
            new.append((ix,iy))
        for ix,iy in newFrontier: #προσθέτει τα καινούρια κελιά στο νέο frontier
          new.append((ix,iy))
        return new

      frontier = GetFrontier(self.grid, self.N, 1, 1) #βρίσκει το αρχικό frontier
      while len(frontier) > 0:  #όσο υπάρχουν κελιά στο frontier
        randomindex = random.randint(0, len(frontier)-1)  #διαλέγουμε ένα στην τύχη
        (fx,fy) =(frontier[randomindex][0], frontier[randomindex][1])
        self.grid[fx,fy]=1  #γίνεται ελεύθερο
        neighbours = GetNeighbours(self.grid, self.N, fx, fy) #βρίσκουμε τα γειτονικά ελεύθερα κελιά
        randNeighbourIndex = random.randint(0, len(neighbours)-1) #διαλέγουμε ένα στην τύχη
        (nx,ny) = (neighbours[randNeighbourIndex][0], neighbours[randNeighbourIndex][1])  #βρίσκου τις συντεταγμένες του γειτονικού κελιού
        (mx,my) = ((fx+nx)//2, (fy+ny)//2) #βρίσκουμε τις συντεταγμένες του ενδιάμεσου κελιού
        self.grid[mx,my]=1  #κάνουμε το ενδιάμεσο κελί ελεύθερο
        newfrontier = GetFrontier(self.grid, self.N, fx, fy) #βρίσκουμε το νέο Frontier
        frontier = mergefrontiers(self.grid, frontier, newfrontier) #ανανεώνουμε το παλιό

      class Node: #μας βοηθάει για να κρατάμε πληροφορία για τα κελιά
        def __init__(self, position):
            self.position = position
            self.distance = 0

      def dfs(startNode, closedList):   #αλγόριθμος DFS
        closedList.append((startNode.position, startNode.distance)) #βάζουμε στην κλειστή λίστα τις συντεταγμένες του κελιού και την απόσταση από τον αρχικό στόχο
        for movePosition in [(0, 1), (0, -1), (1, 0), (-1, 0)]: #για όλα τα πιθανά βήματα
          newPosition = (movePosition[0] + startNode.position[0], movePosition[1] + startNode.position[1])  #βρίσκουμε τη νέα θέση
          if self.grid[newPosition[0], newPosition[1]] == 0:  #αν είναι τοίχος πάμε στο επόμενο πιθανό βήμα
              continue
          nextNode = Node(newPosition)  #αλλιώς δημιουργούμε Node που δείχνει την τωρινή νέα θέση
          nextNode.distance = startNode.distance + 1  #υπολογίζουμε την απόσταση από τον αρχικό στόχο
          flag = 0  #το χρησιμοποιούμε για να ελέγξουμε ότι δεν το έχουμε ξανακοιτάξει
          for position, distance in closedList: #ελέγχουμε αν το έχουμε ξανακοιτάξει
            if position == newPosition: 
              flag = 1  #αν ναι κάνουμε το flag=1
              break
          if flag == 0: #αν δεν έγινε το Flag=1 τότε δεν το έχουμε ξαναπετύχει 
            dfs(nextNode, closedList) #συνεχίζουμε το dfs από τον νέο κελί

      startingNode = Node(self.S)   #αρχικός κόμβος που αναπαρηστά το κελί Αρχή
      startingNode.distance = 0 #απόσταση αρχικοποιείται σε 0
      closedSList = []  #κλειστή λίστα που αφορά την απόσταση από την αρχή
      dfs(startingNode, closedSList)  #dfs με αρχικό κελί την Αρχή
      closedSList.sort()  #sortαρισμα της κλειστής λίστας με βάση της συντεταγμένες
      finishNode = Node(self.F) #αρχικός κόμβος που αναπαρηστά το κελί Τέλος
      finishNode.distance = 0 #απόσταση αρχικοποιείται σε 0
      closedFList = []  #κλειστή λίστα που αφορά την απόσταση από το τέλος
      dfs(finishNode, closedFList)  #dfs με αρχικό κελί το Τέλος
      closedFList.sort() #sortαρισμα της κλειστής λίστας με βάση της συντεταγμένες             

      startnodes = [] #περιοχή αρχής
      finishnodes = []  #περιοχή τέλους
      while len(closedSList) > 0 and len(closedFList) > 0:  #όσο δεν είναι κενές οι κλειστές λίστες
        position1, distance1 = closedSList[0] #πρώτο στοιχείο στην κλειστή λίστα που αφορά αποστάσεις από την αρχή
        position2, distance2 = closedFList[0] #πρώτο στοιχείο στην κλειστή λίστα που αφορά αποστάσεις από το τέλος
        if distance1 >= distance2:  # αν είναι πιο κοντά στο τέλος
          finishnodes.append(position1) #μπαίνει στην περιοχή τέλους
        else:
          startnodes.append(position1)    #μπαίνει στην περιοχή αρχής
        closedSList.pop(0)
        closedFList.pop(0)
      
      var = self.N//2 + 1   #παράμετρος για το πόσα κελιά από εμπόδια θα γίνουν ελεύθερα γαι να δημιουργηθούν περισσότερα μονοπάτια 
      for stpos in startnodes: #διατρέχει την περιοχή αρχής
        if var == 0:  #αν έσπασαν τόσα κελιά όσα και η παράμετρος τότε Break
          break
        for finpos in finishnodes:  #διατρέχει την περιοχή τέλους
          if var == 0:   #αν έσπασαν τόσα κελιά όσα και η παράμετρος τότε Break
            break 
          if (stpos[0] - finpos[0])==2 and stpos[1]==finpos[1] and self.grid[stpos[0]-1, stpos[1]]==0: #αν είναι γειτονικά κελιά και το ενδιάμεσο είναι εμπόδιο
            self.grid[stpos[0]-1, stpos[1]]=1 #κάνε το ενδιάμεσο ελεύθερο
            var -= 1  #μείωσε το μετρητή κατά 1
          elif (finpos[0] - stpos[0])==2 and stpos[1]==finpos[1] and self.grid[stpos[0]+1, stpos[1]]==0:  #αν είναι γειτονικά κελιά και το ενδιάμεσο είναι εμπόδιο
            self.grid[stpos[0]+1, stpos[1]]=1 #κάνε το ενδιάμεσο ελεύθερο
            var -= 1  #μείωσε το μετρητή κατά 1
          elif (stpos[1] - finpos[1])==2 and stpos[0]==finpos[0] and self.grid[stpos[0], stpos[1]-1]==0:  #αν είναι γειτονικά κελιά και το ενδιάμεσο είναι εμπόδιο
            self.grid[stpos[0], stpos[1]-1]=1 #κάνε το ενδιάμεσο ελεύθερο
            var -= 1  #μείωσε το μετρητή κατά 1
          elif (finpos[1] - stpos[1])==2 and stpos[0]==finpos[0] and self.grid[stpos[0], stpos[1]+1]==0:  #αν είναι γειτονικά κελιά και το ενδιάμεσο είναι εμπόδιο
            self.grid[stpos[0], stpos[1]+1]=1 #κάνε το ενδιάμεσο ελεύθερο
            var -= 1  #μείωσε το μετρητή κατά 1
            

          


    def draw_map(self, path=None):
        """ Draws the maze as an image. Considers grid values of 0/False to represent obstacles and
        values of 1/True to represent empty cells, but this can be customized. Obstacles are painted
        black and empty cells are painted white. Starting point is painted green and finish point red.
        Optionally accepts as a parameter a path within the maze which is painted blue. 
        """
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[~self.grid] = [0, 0, 0]
        image[self.grid] = [255, 255, 255]

        # Uncomment the next 2 lines of code to treat 1/True as obstacles (and 0/False as free maze cells)
        # image[self.grid] = [0, 0, 0]
        # image[~self.grid] = [255, 255, 255]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()
````

````
class Pathfinder:
    def __init__(self, maze, c, h):
        """
        maze : Αντικείμενο τύπου Maze (από Μέρος 1)
        c : Συνάρτηση που υπολογίζει την πραγματική απόσταση μεταξύ δύο σημείων
        h : Συνάρτηση που υπολογίζει την ευριστική μεταξύ δύο σημείων
        """
        self.maze = maze
        #self.vis = visualization(self.maze.S, self.maze.F)
        self.path = []
        self.cost = c
        self.heuristic = h


    

    ### Fill the path list with the coordinates of each point in the path from maze.S to maze.F
    ### Your code here
    class Node:   #για να έχουμε τις πληροφορίες κάθε κελιού
        def __init__(self, x, y, isWall):
            self.xposition = x
            self.yposition = y
            self.isWall = isWall
            self.parent = None
            self.g = 0
            self.h = 0
            self.f = 0


    def astar(self, closedList): #συνάρτηση που υλοιποιεί τον Α* παίρνει ως παράμετρο το κλειστό σύνολο
        startNode = self.Node(self.maze.S[0], self.maze.S[1], False)  #αρχικό κελί είναι ελεύθερο
        finishNode = self.Node(self.maze.F[0], self.maze.F[1], False) #τελικό κελί είναι ελεύθερο

        openList = [] #μέτωπο αναζήτησης
        startNode.g = 0 #αρχικό κόστος 0
        startNode.h = self.heuristic([startNode.xposition, startNode.yposition],[finishNode.xposition, finishNode.yposition]) #εύρεση ευριστικής τιμής
        startNode.f = startNode.g + startNode.h #εύρεση συνολικής τιμής
        openList.append(startNode)  #προσθήκη του αρχικού κελιού στο μέτωπο αναζήτησης
        frontier = []
        expanded = []

        while len(openList) > 0:  #όσο το μέτωπο αναζήτησης δεν είναι κενό
 
            currentNode = openList[0] #τωρινός κόμβος ο πρώτος κόμβος στο μέτωπο αναζήτησης
            openList.pop(0) #τον βγάζουμε από το μέτωπο αναζήτησης
            
            frontier.append((currentNode.xposition, currentNode.yposition))


            if currentNode.xposition == finishNode.xposition and currentNode.yposition == finishNode.yposition: #άμα είναι το Τέλος
                while currentNode is not None:  #διατρέχουμε τον γονέα κάθε κόμβου μέχρι να φτάσουμε στην αρχή που έχει πατέρα το None
                    self.path.append((currentNode.xposition, currentNode.yposition))  #βάζουμε τις συντεταγμένες στο μονοπάτι
                    currentNode = currentNode.parent  #πάμε στον πατέρα του
                break #βγαίνουμε από το while loop

            for movePosition in [(0, 1), (0, -1), (1, 0), (-1, 0)]: #για όλες τις πιθανές κατευθύνσεις, τα πιθανά παιδιά
                newPosition = (movePosition[0] + currentNode.xposition, movePosition[1] + currentNode.yposition)  #υπολογισμός νέας θέσης
                if self.maze.grid[newPosition[0], newPosition[1]] == 0: #αν είναι τοίχος πάμε στην επόμενη πιθανή θέση
                    #    print("wall")
                    continue

                childNode = self.Node(newPosition[0], newPosition[1], False) #δημιοργία κόμβου για το παιδί
                childNode.g = self.cost(currentNode.g) #υπολογισμός πραγματικού κόστους
                childNode.parent = currentNode  #κάνουμε γονέα του παιδιού που ελέγχουμε, τον τωρινό κόμβο
                flag = 0  #θα μας βοηθήσει να δούμε αν το έχουμε ξανακοιτάξει
                for index, closedChild in enumerate(closedList):  #ελέγχος αν είναι μέσα στο κλειστό σύνολο
                    if closedChild.xposition == newPosition[0] and closedChild.yposition == newPosition[1]: #αν ανήκει
                        if childNode.g >= closedChild.g: #και έχει χειρότερο κόστος από ότι το αντίστοιχο στο κλειστό σύνολο
                            flag = 1  #το Flag =1 για να μα πει να μην το βάλουμε ξανά στο μέτωπο αναζήτησης
                            break
                        else:
                            closedList.pop(index) #αν βρήκαμε καλύτερη περίπτωση, βγάζουμε τον κόμβο από το κλειστό σύνολο
                if flag == 1:
                    continue  #αν το flag = 1 πάμε στο επόμενο παιδί
                childNode.h = self.heuristic([childNode.xposition, childNode.yposition], [finishNode.xposition, finishNode.yposition]) #υπολογισμός ευριστικής
                childNode.f = childNode.g + childNode.h #υπολογισμός συνολικού κόστους
                flag = 0 #θα μας βοηθήσει να κρατήσουμε μόνο την καλύτερη περίπτωση
                if len(openList) == 0:  #αν το μέτωπο αναζήτησης είναι άδειο τότε το βάζουμε κατευθείαν
                    openList.append(childNode)
                    frontier.append((childNode.xposition, childNode.yposition))
                else: 
                    for index, openNode in enumerate(openList): #συγκρίνουμε τα κόστοι και βλέπουμε αν υπάρχει ήδη
                        if childNode.xposition == openNode.xposition and childNode.yposition == openNode.yposition and childNode.f >= openNode.f:  #αν υπάρχει ήφη και τώρα έχει χειρότερο κόστος
                            flag = 1  #κάνουμε το Flag=1 για να μην το προσθέσουμε στο μέτωπο αναζήτησης
                            break
                        if childNode.f >= openNode.f:  #συγκρίνουμε τις ευριστικές για να μπει στη σωστή κατάταξη
                            continue #αν είναι χειρότερο τότε προχωράμε
                        else:
                            openList.insert(index, childNode) #αν είναι καλύτερο το προσθέτουμε στη θέση που πρέπει να διατηρηθεί η διάταξη
                            frontier.append((childNode.xposition, childNode.yposition))
                            flag = 1
                            break
                    if flag == 0: #αν δεν το βρήκαμε στο μέτωπο αναζήτηση και είχε χειρότερη ευριστική από όλους τους άλλους κόμβους
                        openList.append(childNode)  #το προσθέτουμε στο τέλος
                        frontier.append((childNode.xposition, childNode.yposition))
            
            closedList.append(currentNode)  #βάζουμε τον τωρινό κόμβο στο κλειστό σύνολο
            expanded.append((currentNode.xposition, currentNode.yposition))
            #self.vis.draw_step(self.maze, frontier, expanded)
                    

    def get_path_expan(self): #συνάρτηση που θα χρησιμοποιήσουμε για τα στατιστικά
      closedList = []
      self.astar(closedList)
      return len(self.path), len(closedList)  #επιστρέφει το μήκος του μονοπατιού και το πλήθος των expanded nodes


    def get_path(self):
        closedList = []
        self.astar(closedList)
        return self.path
````

* $g(n)=0$ και $h(n)=\{\text{manhattan}(n),\text{euclidean}(n)\}$. Είναι ο αλγόριθμος Best First, κοιτάει μόνο την ευριστική. Δεν μπορεί να βρει το βέλτιστο μονπάτι καθώς δεν γνωρίζει το κόστος της πραγματικής απόστασης και γιατί μόλις βρει την πρώτη λύση σταματάει
* $g(n)=1$ και $h(n)=0$. Είναι ο αλγόριθμος Dijkstra. Βρίσκει πάντα τη βέλτιστη λύση αφού υποολγίζει τα κόστη όλων των πιθανών μονοπατιών και επιστρέφει εκείνο με το μικρότερο κόστος (το βέλτιστο)
* $g(n)=1$ και $h(n)=\{\text{manhattan}(n),\text{euclidean}(n)\}$. Είναι ο αλγόριθμος A*. Βρίσκει πάμτα τη βέλτιστη λύση καθώς η ευριστικές που χρησιμοποιούνται δεν υποεκτιμούν ποτέ το κόστος επίτευξης του στόχου.

````
def find_points(image_array, pixels):
    points = []
    for grid_i, i in enumerate(range (10, image_array.shape[0]-10, 20)):
        for grid_j, j in enumerate(range (10, image_array.shape[1]-10, 20)):
            if np.array_equal(image_array[i][j], pixels):
                points.append([grid_i+1, grid_j+1])
    return points

def load_maze(fname='/content/drawing.png'):
    N = 17
    image = PIL.Image.open(fname)
    image_array = np.round(np.array(image)/255)

    start_x, start_y = find_points(image_array, [0,1,0,1])[0]
    end_x, end_y = find_points(image_array, [1, 0, 0, 1])[0]
    walls = find_points(image_array, [0, 0, 0, 1])
    new_grid = Maze(N, (start_x, start_y), (end_x, end_y))

    for i in range(N):
        for j in range(N):
            if [i, j] in walls:
                new_grid.grid[i,j] = 0
            else:
                new_grid.grid[i, j] = 1

    g = new_grid
    g.draw_map()
    return g
````
